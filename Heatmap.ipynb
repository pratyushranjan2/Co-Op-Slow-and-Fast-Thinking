```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
```

```python
df1 = pd.read_csv("heat_map_sys1.csv")
df2 = pd.read_csv("heat_map_sys2.csv")
```

```python
df1['x,y'] = df1[['x', 'y']].apply(lambda x: ','.join(x.map(str)), axis=1)
df2['x,y'] = df2[['x', 'y']].apply(lambda x: ','.join(x.map(str)), axis=1)
```

```python
df1['freq'] = df1.groupby('x,y')['x,y'].transform('count')
df1 = df1.drop_duplicates()
df2['freq'] = df2.groupby('x,y')['x,y'].transform('count')
df2 = df2.drop_duplicates()
```

```python
fig = plt.gcf()
fig.set_size_inches(20, 5)
plt.scatter(df1['x'], df1['y'], alpha=.01, s=200,c = 'black')
walls_x = [3,4,8,9,12,13,17,18,3,4,6,15,17,18,3,4,6,8,9,10,11,12,13,15,17,18,8,13]
walls_y = [4,4,4,4,4,4,4,4,3,3,3,3,3,3,2,2,2,2,2,2,2,2,2,2,2,2,5,5]
print (len(walls_x))
print (len(walls_y))
plt.scatter(walls_x,walls_y,s = 800, marker = 's')
plt.yticks(range(1, 6))
plt.xticks(range(1, 21))
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
fig = plt.gcf()
fig.set_size_inches(20, 5)
# s = [20*2**n for n in range(len(x))]
plt.scatter(df1['x'], df1['y'],c = 'black' , s = df1['freq']*2,label='System 1') 
plt.scatter(df2['x'], df2['y'],c = 'red' , s = df2['freq']*2,label='System 2') 
# plt.scatter(df2['x'], df2['y'], alpha=.08, s=200,c = 'black')
walls_x = [3,4,8,9,12,13,17,18,3,4,6,15,17,18,3,4,6,8,9,10,11,12,13,15,17,18,8,13]
walls_y = [4,4,4,4,4,4,4,4,3,3,3,3,3,3,2,2,2,2,2,2,2,2,2,2,2,2,5,5]
print (len(walls_x))
print (len(walls_y))
plt.scatter(walls_x,walls_y,s = 800, marker = 's')
plt.yticks(range(1, 6))
plt.xticks(range(1, 21))
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),ncol=3, fancybox=True, shadow=True)
plt.grid()
plt.show()
```
